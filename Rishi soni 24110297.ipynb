```python
#Question1
natural=list(range(1,21))
print("List of 20 natural numbers: ",natural)
b=list(filter(lambda a: a%2==0, natural))
print("Sum of all even numbers in the list: ",sum(b))
c=list(filter(lambda a: a%2!=0, natural))
n=1
for i in c:
    n=n*i
print("Product of all odd numbers in the list: ",n)
mean=(sum(natural))/(len(natural))
print("mean: ",mean)
median= (natural[9] + natural[10])/2
print("median: ",median)
count=0
for i in natural:
    count=count+ (i-mean)**2
import math
SD= math.sqrt(count/(len(natural)))
print("Standard deviation: ",SD)
```

```python
#Question3
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)
list1=[]
x=int(input("Enter number:"))
for i in range(x):
    res=fib(i)
    list1.append(res)
print(list1)
import matplotlib.pyplot as plt

import numpy as np

plt.plot(range(len(list1)),list1, marker='o')
plt.show()
```

```python
#Question21
list1=[2,14,2,6,6,82,34,68,2,10,14,68,2,82,10]
#Mean
mean= (sum(list1))/len(list1)
print('Mean:',mean)
#Median
list1.sort()
if (len(list1))%2==0:
    a=(len(list1))//2
    b=(list1[a]+list1[a-1])//2
    print('Median:',b)
else:
    c=((len(list1))-1)//2
    d=list1[c]
    print('Median:',d)
#Mode
x=0
y=0
for i in list1:
    if list1.count(i)>x:
        x=list1.count(i)
        y=i
print('Mode:',y)
#Standard deviation
import math
sum1=0
for i in list1:
    sum1=sum1+((i-mean)**2)
val= sum1/len(list1)
SD= math.sqrt(val)
print('Standard deviation:',SD)

import matplotlib.pyplot as plt 
plt.boxplot(list1)
plt.show()
```
